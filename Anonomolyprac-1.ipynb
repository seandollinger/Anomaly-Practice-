```python
# import library
import numpy as np 
from sklearn.ensemble import IsolationForest 
import matplotlib.pyplot as plt 
```

```python
#reshape to fit model input format 

data = np.random.normal(scale= 0.5, size=1000) 

#reshape to fit model input format 

data = data.reshape(-1, 1) 

plt.plot(range(len(data)), data)
plt.show()
```

```python
# create Isolation Forest Model 

clf = IsolationForest(contamination=0.5) #contamination is the proprtion 5% 

# Fit the model and predict anomalies 

predictions = clf.fit_predict(data) 

# Anomalies are labeled as -1, normal points are labled as 1 

anomoly_indices = np.where(predictions == -1) [0] 

plt.plot(range(len(data)), data) 
plt.scatter (anomoly_indices, data[anomoly_indices], c = 'r') 
plt.show
```
